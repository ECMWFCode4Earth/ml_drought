```python
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

fig, ax = plt.subplots(figsize=(12, 8))
world[world.name == 'Kenya'].plot(ax=ax, color='#D8D8D8')

ax = plt.gca()
ax.axis('off');
```

```python
world.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
world[world.continent == 'Africa'].plot(ax=ax, color='#D8D8D8')

ax = plt.gca()
ax.axis('off');
```

```python
countries = ['Ethiopia', 'Uganda', 'Kenya', 'Somalia', 'Djibouti', 'Eritrea', 'Somaliland', 'Rwanda', 'Burundi', 'Sudan', 'S. Sudan']

fig, ax = plt.subplots(figsize=(12, 8))
world[np.isin(world.name, countries)].plot(ax=ax, color='#D8D8D8')

ax = plt.gca()
ax.axis('off');
```
